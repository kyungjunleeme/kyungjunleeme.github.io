```python
import json
import pandas as pd
import numpy as np

'''bbox[0] = 'ymin'
   bbox[1] = 'xmin'
   bbox[2] = 'ymax'
   bbox[3] = 'xmax'
# 0: ymin, 1: xmin, 2: ymax, 3: xmax'''
colum_list = ['json_source', 'filename', 'obj_list', 'obj_sum_cnt']
# ans_obj_list=[]

object_list = []
# row_list = []

# per_file_per_object_info_list = []

df_1 = pd.DataFrame(columns=colum_list)


with open('1106_object_list.json') as json_file:
    json_data = json.load(json_file)
    file_name_list = list(json_data)
    
    for filename in file_name_list:
        print(filename)
        object_list[:]=[]
        print("object_list 조사",object_list)
        file_info = json_data[filename]
        '''['0', '1', '2', '3', '4']'''
        
        obj_cnt_list = list(file_info)
        obj_cnt = len(obj_cnt_list)
        print('여기서', obj_cnt_list)
        if not obj_cnt_list:
            print("List 가 비었다")
            df_1 = df_1.append({
                "json_source": '1106_object_list.json',
                "filename": filename,
                "obj_sum_cnt": "0",
                "obj_list": "Null"
            }, ignore_index=True)
        else:
            print("List가 비지 않았다")
            for obj_idx in obj_cnt_list[:]:
                try:
                    pop_ele = obj_cnt_list.pop(len(obj_idx)-1)
                    object_list.append(file_info[pop_ele]['name'])
                    if not obj_cnt_list:
                        print(filename, obj_cnt, object_list)
                        np_mat = np.array(object_list)
                        print(np_mat)
                        df_1 = df_1.append({
                            "json_source": '1106_object_list.json',
                            "filename": filename,
                            "obj_sum_cnt":  obj_cnt,
                            "obj_list":  np_mat
                        }, ignore_index=True)
                    else:
                        print('obj_cnt_list에서 꺼낼 것이 있다')
                except:
                    ''' pop_ele에 원소 없으면 오류 발생할테니 '''
                    continue
```
