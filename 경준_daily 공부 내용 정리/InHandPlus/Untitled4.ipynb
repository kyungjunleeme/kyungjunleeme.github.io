```python
import json
import pandas as pd
import numpy as np
import pandas as pd
import collections
from collections import Counter,defaultdict

answer_video_list = ["200406_175250_null.mp4", "200406_175746_null.mp4", "200406_180239_null.mp4"]
answer_data = pd.read_csv("C:\\Users\\godof\\Desktop\\answer.csv", names=['video', 'frame_jpg'])
answer_df_to_list = []
answer_dict = {}

for i in range(len(answer_data)):
    answer_df_to_list.append(list(answer_data.iloc[i]))


for video, frame_jpg in answer_df_to_list:
    if video not in answer_dict:
        answer_dict[video] = []
    answer_dict[video].append(frame_jpg)

sum_result_detection = []
per_video_result_detection_depth_1 = {}
per_video_result_detection_depth_2 = {}
sum_result_count = []

sum_result_detection = []
sum_result_count = []

sum_all_data = list()
all_data = defaultdict(dict)
video_data = defaultdict(list)

'''
json 파일이 들어 있는 폴더 내에서 checkpoint_num 만 분리해서 잘라낸 다음 확인 추출 할 예정
checkpoint_num_list = [23902,24472,25043,25614,26185]
'''
checkpoint_num_list = [23902,24472,25043,25614,26185]


# 200406_175746_null.jpg
for answer_video in answer_video_list:
    for checkpoint in checkpoint_num_list:
        typ_chged_checkpoint = str(checkpoint)
        with open('D:\\[DB_Review]\\[20200409 DB_Review_pullout_main_v2_try2]\\03. Result\\nipa_json\\{}\\{}_{}.json'.format(answer_video.split(".")[0], answer_video.split(".")[0], typ_chged_checkpoint)) as json_file:
            count = 0
            jpg_num = 0
            json_data = json.load(json_file)
            for idx, frame_info in enumerate(json_data):
                try:
                    # object 가 여러 개 나올 경우 순서는 반영함 
                    # 중복은 count 에서는 고려되었는데, 개별 검증에서는 반영된건지 확인 필요
                    for i in range(5):                    
                        if frame_info[i]['name'] == 'pullout_main_v2':
                            count = count + 1
                            sum_result_detection.append([typ_chged_checkpoint, answer_video, "{}_{}.jpg".format(answer_video.split(".")[0], idx+1)])



                            '''
                            sum_result_detection.append(["{}".format(answer_video),
                            answer_video.split(".")[0]+'_'+typ_chged_checkpoint,
                            "{}_{}.jpg".format(answer_video.split(".")[0], idx+1)])

                            '''


                        else:
                            continue
                except:
                    pass
        sum_result_count.append([answer_video.split(".")[0]+'_'+typ_chged_checkpoint, count])
    else:
        continue



for idx, (check_num, video_name, frame_jpg) in enumerate(sum_result_detection):
    if idx == 0:
        if check_num not in all_data:
            all_data[check_num]
            all_data[check_num] = video_data
            video_data[video_name].append(frame_jpg)
        else:
            continue
    else:
        if check_num not in all_data:
            sum_all_data.append(all_data)
            all_data = defaultdict(dict)
            video_data = defaultdict(list)
            if sum_result_detection[idx-1][1] == sum_result_detection[idx][1]:
                all_data[check_num]
                all_data[check_num] = video_data
                video_data[sum_result_detection[idx][1]].append(sum_result_detection[idx][2])
            else:
                all_data[check_num]
                all_data[check_num] = video_data
                video_data[sum_result_detection[idx][1]].append(sum_result_detection[idx][2])
        else:
            if idx == len(sum_result_detection)-1:
                video_data[sum_result_detection[idx][1]].append(sum_result_detection[idx][2])
                sum_all_data.append(all_data)
            else:
                video_data[sum_result_detection[idx][1]].append(sum_result_detection[idx][2])
```
