```python
import json
import pandas as pd
import numpy as np
import itertools
```

```python
ans_df = pd.read_csv("C:\\Users\\godof\\Documents\\Data_Model_Management\\test_image_output\\20191114.csv")
```

```python
ans_df = ans_df.drop(['width', 'height', 'obj_idx', 'ymin', 'xmin', 'ymax', 'xmax'], axis=1)
'''
특정 열을 날리고
'''
dict_ans_df = ans_df.to_dict
lst_dict_ans_df = list(dict_ans_df('records'))

colum_list = ['csv_source', 'filename', 'obj_list', 'obj_sum_cnt']
df_3 = pd.DataFrame(columns=colum_list)

lst_dict_ans_df = sorted(lst_dict_ans_df, key=lambda x: x['filename'])
for i, v in itertools.groupby(lst_dict_ans_df, lambda i:i['filename']):
    save_lst = list(v)
    obj_lst=[]
    if not obj_lst: 
        for cnt in range(0, len(save_lst)):
            obj_lst.append(save_lst[cnt]['obj_name'])
            np_mat = np.array(obj_lst)
            obj_sum_cnt = len(np_mat)
            if len(save_lst) - cnt==1:
                df_3 = df_3.append({
                    'csv_source': '20191114.csv',
                    'filename': i,
                    'obj_list': np_mat,
                    'obj_sum_cnt': obj_sum_cnt
                }, ignore_index=True)
            else:
                print("아직 남았다")
    else:
        print('비었다')
```
