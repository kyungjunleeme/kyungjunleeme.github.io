```python
import json
import pandas as pd
import numpy as np

ans_df = pd.read_csv("C:\\Users\\godof\\Documents\\Data_Model_Management\\test_image_output\\20191113.csv")
```

```python
ans_df
```

```python
ans_df['per_obj_cnt'] = np.where(ans_df['obj_idx'] != 'Null' , '1', '0')
ans_df['model'] = 'answer_120'+'_'
ans_df['csv_source'] = '20191113.csv'
ans_df['result_per_obj_id_int'] = ans_df.groupby(['filename','per_obj_cnt']).cumcount()+1
ans_df['result_per_obj_id_str'] = ans_df['result_per_obj_id_int']

'''
apply 쓰기 위해 int-> str 형 변환
'''

ans_df['result_per_obj_id_str'] = ans_df['result_per_obj_id_str'].apply(str)

'''
str 형태만 결합할 수 있음, int형태는 결합 안 됨
https://www.geeksforgeeks.org/create-a-new-column-in-pandas-dataframe-based-on-the-existing-columns/
(참고)
cols = list(df.columns.values)
'''
ans_df['filename']= ans_df['filename']+'_'
ans_df['ans_per_obj_pk'] = ans_df.apply(lambda row: row.model + row.filename+ row.result_per_obj_id_str , axis=1)
ans_df = ans_df.reindex(columns=['csv_source', 'ans_per_obj_pk', 'model', 'filename', 'result_per_obj_id_str', 'obj_name', 'obj_idx', 'score', 'xmin', 'ymin', 'xmax', 'ymax', 'width', 'height', 'per_obj_cnt', 'result_per_obj_id_int'])
ans_df['score']=1.0
```
