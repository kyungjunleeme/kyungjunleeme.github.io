```python
# -*- coding: utf-8 -*-
import os


path = "G:\\ \\InHandPlus\\03. AI \\03. \\\\_\\medi_001\\"
file_list = os.listdir(path)


file_list_jpg = [jpg_file for jpg_file in file_list if jpg_file.endswith(".jpg")]
file_list_xml = [xml_file for xml_file in file_list if xml_file.endswith(".xml")]

for i in range(len(file_list_jpg)):
    file_list_jpg.sort()
    file_list_xml.sort()
    if file_list_jpg[i].split('_')[0] == file_list_xml[i].split('_')[0]:
        only_jpg_name = file_list_jpg[i]
        real_only_jpg_name = only_jpg_name.split('.')[0]
        only_xml_name = file_list_xml[i]
        real_only_xml_name = only_xml_name.split('.')[0]
        try:
            os.rename(path+only_xml_name, path+only_xml_name.replace(real_only_xml_name,real_only_jpg_name))
        except:
            print("")

```

```python
# -*- coding: utf-8 -*-
import os
import xml.etree.ElementTree as et

path = "G:\\ \\InHandPlus\\03. AI \\03. \\\\_\\medi_001\\"
os.chdir(path)
file_list = os.listdir(path)


file_list_jpg = [jpg_file for jpg_file in file_list if jpg_file.endswith(".jpg")]
file_list_xml = [xml_file for xml_file in file_list if xml_file.endswith(".xml")]

for i in range(len(file_list_jpg)):
    file_list_jpg.sort()
    file_list_xml.sort()
    if file_list_jpg[i].split('_')[0] == file_list_xml[i].split('_')[0]:
        tree = et.parse(file_list_xml[i])
        tree.find('./folder').text = 'medi_001'
        tree.find('./filename').text = file_list_jpg[i]
        tree.find('./path').text = path+file_list_jpg[i]
        tree.write(file_list_xml[i], encoding="utf-8", xml_declaration=True)
    else:
        print("")
```

```python
# -*- coding: utf-8 -*-
import os
import shutil



path = input()
merge_path = input()
folder_list = []

object_list = ['bottle', 'cap', 'cup', 'face', 'grab_bottle_hand', 'hand', 'mouth', 'open_bottle', 'open_mouth', 'pill']
merge_file_list = []


folder_files = os.listdir(path)
for fd_nm in folder_files:
    folder_list.append(fd_nm)

    
    
for folder_name in folder_list[:len(folder_list)]:
    print(folder_name)
    path_pls_folder = path+'\\'+folder_name
    '''D:\test_1023\0827_3_checked'''
    for obj in object_list[:len(object_list)]:
        path_pls_folder_obj = path_pls_folder+'\\'+obj
        '''D:\test_1023\0827_3_checked\bottle'''
        path_pls_folder_obj_files = os.listdir(path_pls_folder_obj)
        for path_pls_folder_obj_dst in path_pls_folder_obj_files:
            '''0000_0827_3_1_bottle.jpg'''
            merge_file_list.append(path_pls_folder_obj_dst)
            shutil.copy(path_pls_folder_obj+'\\'+path_pls_folder_obj_dst, merge_path)
            
print(len(merge_file_list))        
```
