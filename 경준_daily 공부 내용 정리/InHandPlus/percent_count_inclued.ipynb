```python
import csv
import collections
import argparse



final_list=[]
img_name = 0
obj_name = 3
score = 5
new_obj = dict()
obj = dict()

f = open('C:\\Users\\godof\\Desktop\\kyungjun\\company\\detection_result\\data_0302\\2020_03_03_Test_LHW_01.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
for line in rdr:
    # line 은 # ['0141_0827_1.jpg', '1280', '720', 'cup', '200', '1', '1280', '720']
    # 
    if line[obj_name] in obj:
        obj[line[obj_name]] += 1
    else:
        # obj[line[3]] = 1
        # obj['cup'] = 10
        # obj['bottle'] = 20
        # obj['cap'] = 30
        obj[line[obj_name]] = 1

f.close()
obj.keys()
# return의 형태가 dictionary
del obj['obj_name']
print(obj)

f_1 = open('C:\\Users\\godof\\Desktop\\kyungjun\\company\\detection_result\\data_0302\\2020_03_03_Test_LHW_01.csv', 'r', encoding='utf-8')
rdr_1 = csv.reader(f_1)
for line_1 in rdr_1:
    if not line_1[score] in new_obj:
        # new_obj[line_1[3]] = line_1[5]
        if line_1[score] == 'score':
        # if line_1[obj_name] == 'Null' or line_1[score] == 'score' or 0:
            pass
        else:
            new_obj[line_1[obj_name]] = float(line_1[score])
            ## dictionary 인 new_obj 를 ditonary로 한번 더 감쌈
            final_list.append(dict(new_obj).copy())
            del new_obj[line_1[obj_name]]
    else:
        pass
```
