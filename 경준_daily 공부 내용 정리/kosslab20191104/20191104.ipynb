```python
import requests
from bs4 import BeautifulSoup
url = "https://finance.naver.com"
res = requests.get(url)

soup = BeautifulSoup(res.text, "html.parser")
tbody = soup.find("tbody", id="_topItems1")
a_list = tbody.find_all("tr")
for item in a_list:
    print(item.th.text, end="\t")
    sub_td = item.find_all("td")
    for td_item in sub_td:
        print(td_item.text,end='\t')
    print()
```
